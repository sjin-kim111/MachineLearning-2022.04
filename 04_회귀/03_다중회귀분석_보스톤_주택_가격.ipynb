```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- 데이터 전처리

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['Price'] = boston.target
df.head()
```

- Feature별 산점도와 선형회귀선

```python
# CRIM Feature 확인
sns.regplot(x='CRIM', y='Price', data=df);
```

```python
# 나머지 12개 Feature 확인
_, axs = plt.subplots(3, 4, figsize=(16, 12))
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i // 4, i % 4
    sns.regplot(x=feature, y='Price', data=df, ax=axs[row][col])
```

- Feature별 결정 계수(R squared)

```python
from sklearn.linear_model import LinearRegression
```

```python
for feature in df.columns[:-1]:
    lr = LinearRegression()
    X = df[feature].values.reshape(-1, 1)
    lr.fit(X, boston.target)
    score = lr.score(X, boston.target)
    print(f'{feature}:\t{score:.4f}')
```

- 다중선형회귀
<pre>
회귀식
y = -1.18 * CRIM + 4.72 * ZN + 4.93 * INDUS + ... -5.27 * LSTAT + 30.95
</pre>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.1, random_state=2022)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# R squared
lr.score(X_train, y_train)
```

```python
# 계수, Weight
lr.coef_
```

```python
# 절편, bias
lr.intercept_
```

- 테스트 데이터에 적용

```python
X_test[0], y_test[0]
```

```python
# Weight와 bias를 이용하여 직접 구한 값
np.dot(lr.coef_, X_test[0]) + lr.intercept_
```

```python
# predict method
lr.predict(X_test[0].reshape(1, -1))
```

```python
# 10개의 테스트 데이터에 적용
for i in range(10):
    pred1 = np.dot(lr.coef_, X_test[i]) + lr.intercept_
    pred2 = lr.predict(X_test[i].reshape(1, -1))
    print(f'실제값: {y_test[i]},\t직접 계산 예측값: {pred1:.4f},\tLR 예측값: {pred2[0]:.4f}')
```

### 회귀는 꼭 선형회귀를 써야하는가?
- Linear Regression, Decision Tree, SVM, Random Forest, XGBoost

```python
from sklearn.metrics import r2_score, mean_squared_error
```

1. Linear Regression

```python
pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, pred_lr)
mse_lr = mean_squared_error(y_test, pred_lr)
```

2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=2022)
dtr.fit(X_train, y_train)
pred_dt = dtr.predict(X_test)
r2_dt = r2_score(y_test, pred_dt)
mse_dt = mean_squared_error(y_test, pred_dt)
```

3. SVM

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
pred_sv = svr.predict(X_test)
r2_sv = r2_score(y_test, pred_sv)
mse_sv = mean_squared_error(y_test, pred_sv)
```

4. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=2022)
rfr.fit(X_train, y_train)
pred_rf = rfr.predict(X_test)
r2_rf = r2_score(y_test, pred_rf)
mse_rf = mean_squared_error(y_test, pred_rf)
```

5. XGBoost

```python
from xgboost import XGBRegressor
xgr = XGBRegressor(random_state=2022)
xgr.fit(X_train, y_train)
pred_xg = xgr.predict(X_test)
r2_xg = r2_score(y_test, pred_xg)
mse_xg = mean_squared_error(y_test, pred_xg)
```

6. 각 회귀분석의 R2 score, MSE 비교

```python
print(f'LR\tDT\tSVM\tRF\tXG')
print(f'{r2_lr:.4f}\t{r2_dt:.4f}\t{r2_sv:.4f}\t{r2_rf:.4f}\t{r2_xg:.4f}\t')
print(f'{mse_lr:.4f}\t{mse_dt:.4f}\t{mse_sv:.4f}\t{mse_rf:.4f}\t{mse_xg:.4f}\t')
```

```python
df = pd.DataFrame({'y_test': y_test, 'LR': pred_lr, 'DT': pred_dt, 'SVM': pred_sv, 'RF': pred_rf, 'XG': pred_xg})
df.head()
```
